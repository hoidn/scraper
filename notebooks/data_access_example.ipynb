```python
diffNASDAQ = np.zeros((1,n.shape[0]))
diffNASDJIA = np.zeros((1,n.shape[0]))
diffDJIA = np.zeros((1,n.shape[0]))
diffDJIANAS = np.zeros((1,n.shape[0]))
diffGOLD = np.zeros((1,n.shape[0]))
```

```python
for i in range (0,n.shape[0]-2):
    diff1 = (n[i+1,0]-n[i,0])/(n[i,0])
    diff2 = (n[i+1,1]-n[i,1])/(n[i,1])
    diff3 = (n[i+1,2]-n[i,2])/(n[i,2])
    if abs(diff2) > 0.01:
        diffNASDAQ[0,i] = diff2 
        diffNASDJIA[0,i] = (n[i+1,0]-n[i,0])/(n[i,0])
    if abs(diff1) > 0.01:
        diffDJIA[0,i] = diff1 
        diffDJIANAS[0,i] = (n[i+1,1]-n[i,1])/(n[i,1])
    if abs(diff3) > 0.01:
        diffGOLD[0,i] = diff3
```

```python
t = np.arange(0,242)
```

```python
plt.plot(t,diffNASDAQ[0,:],'r',t,diffNASDJIA[0,:],'b')
plt.show()
```

```python
plt.plot(t,diffNASDAQ[0,:],'r',t,diffDJIA[0,:],'b')
plt.show()
```

```python
plt.plot(t,diffNASDAQ[0,:],'r',t,diffGOLD[0,:],'b')
plt.show()
```
